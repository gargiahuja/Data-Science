```python
final_df=pd.concat([cat_col_encoded,num_col],axis=1)
```

```python
x=final_df.drop("income",axis=1)
y=final_df["income"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
#standard scaler yeh data ko normalix=ze krne k kaam ata h
#yeh data k mean ko 0 krdeta h or standard deviation ko 1 kr dete h
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
#SVC apply krna padega kyoki classification problem h
#SVr for regression
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train_scaled,y_train)
```

```python
svc.score(x_test_scaled,y_test)
```

```python
svc.score(x_train_scaled,y_train)
```

```python
#kaise hum hyperparameter tuning krte h
#kaise hum parameters ko apne according cahnge kr sakte h
from sklearn.model_selection import GridSearchCV
grid={
    "C":[0.01,0.1,1,10],
    "kernel":["linaer",'rbf','play','sigmoid'],
    "degree":[1,3,5,7],
    "gamma":[0.01,1]
}
svm=SVC()
svm_cv=GridSearchCV(svm,grid,cv=5)
svm_cv.fit(x_train_scaled,y_train)
```
