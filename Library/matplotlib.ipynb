```python
x=[1,2,3,4,5]
y=[1,10,9,30,25]
plt.plot(x,y,marker='o',ms=10,mfc="pink",mec="black",color="black",lw=5,ls="--",alpha=1,label="lineplot")
plt.title("plotting")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```

```python
'''
| Marker | Description   |
| ------ | ------------- |
| `'o'`  | Circle        |
| `'s'`  | Square        |
| `'^'`  | Triangle up   |
| `'v'`  | Triangle down |
| `'D'`  | Diamond       |
| `'x'`  | X             |
| `'+'`  | Plus          |
| `'.'`  | Point         |
| `'*'`  | Star          |
| `'p'`  | Pentagon      |
| `'h'`  | Hexagon1      |
| `'H'`  | Hexagon2      |
'''
```

```python
'''
| Style Code    | Line Style    | Example                          |
| ------------- | ------------- | -------------------------------- |
| `'-'`         | Solid line    | `plt.plot(x, y, linestyle='-')`  |
| `'--'`        | Dashed line   | `plt.plot(x, y, linestyle='--')` |
| `'-.'`        | Dash-dot line | `plt.plot(x, y, linestyle='-.')` |
| `':'`         | Dotted line   | `plt.plot(x, y, linestyle=':')`  |
| `''` or `' '` | No line       | Only markers will show           |
'''
```

```python
x=np.linspace(0,10,30)
import warnings
warnings.filterwarnings("ignore")

plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.title("Sin graph")

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.title("Cos graph")

plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.title("tan graph")

plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title("log graph")
plt.show()

```
