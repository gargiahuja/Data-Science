```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#sns k andar humare paas preloaded dataset hote h
df=sns.load_dataset("tips")
df.head()
```

```python
sns.lineplot(df["total_bill"])
sns.scatterplot(df["total_bill"])
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
#data =df-->data source batana
plt.title("Scatterplot")
plt.show()
```

```python
sns.histplot(df["total_bill"],color="red")
plt.show()
plt.hist(df["total_bill"],color="red")
plt.show()
#sns wala is more enhanced jabki parameters are same
```
